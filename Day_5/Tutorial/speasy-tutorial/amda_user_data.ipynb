# Getting AMDA user data
AMDA allows its users to define their own parameters through their web interface. `speasy` can be configured to access these parameters.

```python
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
```

Create a derived parameter in AMDA :



Now that the derived parameter has been created in AMDA we can download it using `speasy`.

```python
import speasy as spz
```

Configure the username and password.

WARNING : users credentials are added to the global configuration file. These credentials will be used in all subsequent `speasy` instances. When using `speasy` to retrieve private parameters on an account other then yours, make sure to reset the `username` and `password`. 

```python
from speasy import config
config.amda_username.set('pyhc')
config.amda_password.set('pyhc')
```

Retrieve the list of user parameters. Each entry is a `ParameterIndex` object that can be passed to `speasy.amda.get_user_parameter`.

Private data retrieval with : 
- `speasy.amda.list_user_parameters`
- `speasy.amda.list_user_timetables`
- `speasy.amda.list_user_catalogs`

```python
l=spz.amda.list_user_parameters()
l
```

Get the user parameter data between the first and second of April 2022.

```python
p=spz.amda.get_user_parameter(l[1],"2022-04-01","2022-04-02")
p.plot()
```
